```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

train = pd.read_csv('/content/mlatmisis/homeworks/hw4/mlcourse-dota2-win-prediction/train_features.csv')
X = train.drop("match_id_hash",axis=1)
tg = pd.read_csv('/content/mlatmisis/homeworks/hw4/mlcourse-dota2-win-prediction/train_targets.csv')
y = tg['radiant_win'].astype(int)


X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,train_size=0.8)



```

```python
import optuna
from sklearn.metrics import roc_auc_score
study = optuna.create_study(direction='maximize')

def f(trial):
    params = {
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'iterations': trial.suggest_int('iterations', 100, 1000),
        'depth': trial.suggest_int('depth', 4, 10),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1e-3, 10),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bylevel': trial.suggest_float('colsample_bylevel', 0.6, 1.0),
        'random_strength': trial.suggest_float('random_strength', 1e-3, 10),
        'verbose': False
    }
    model = CatBoostClassifier(**params)
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    return roc_auc_score(y_test,pred)


study.optimize(f,n_trials=10)

params_c = study.best_params
```

```python
from sklearn.utils import resample

x_params = {'learning_rate': 0.02246671055156704, 'n_estimators': 810, 'max_depth': 4, 'subsample': 0.7411841892465768, 'colsample_bytree': 0.8312577844942544, 'reg_alpha': 4.899293892143854, 'reg_lambda': 2.4282440874351656, 'min_child_weight': 4.626957926761577}
c_params =  {'learning_rate': 0.11836966434805457, 'iterations': 292, 'depth': 4, 'l2_leaf_reg': 2.4849387106988776, 'subsample': 0.8109775692040708, 'colsample_bylevel': 0.899877653572926, 'random_strength': 1.211714320668008}
l_params = {'learning_rate': 0.06956885369432955, 'n_estimators': 704, 'subsample': 0.9581338515408168, 'colsample_bytree': 0.6511570915494759, 'reg_alpha': 0.528378809937175, 'min_child_weight': 8.436287160292284}

#cat = CatBoostClassifier(**c_params)
#xgb = XGBClassifier(**x_params)
#lb = LGBMClassifier(**l_params)

#cat.fit(X_train,y_train)
#xgb.fit(X_train,y_train)
#lb.fit(X_train,y_train)
#
#cat_p = cat.predict_proba(X_test,y_test)
#xgb = xgb.predict_proba(X_test,y_test)
#cat_p = cat.predict_proba(X_test,y_test)

n_estimators = 10
models = {
    "XGBoost": [XGBClassifier(**x_params) for _ in range(n_estimators)],
    "LightGBM": [LGBMClassifier(**l_params) for _ in range(n_estimators)],
    "CatBoost": [CatBoostClassifier(**c_params) for _ in range(n_estimators)]
}

for model_name, model_list in models.items():
    for i in range(n_estimators):
        X_boot, y_boot = resample(X_train, y_train, replace=True)
        model_list[i].fit(X_boot, y_boot)
def predict_proba_bagging(models, X):
    proba = np.zeros((X.shape[0], len(np.unique(y_train))))
    for model in models:
        proba += model.predict_proba(X)
    return proba / len(models)


xgb_proba = predict_proba_bagging(models["XGBoost"], X_test)
lgbm_proba = predict_proba_bagging(models["LightGBM"], X_test)
cat_proba = predict_proba_bagging(models["CatBoost"], X_test)


final_proba = (xgb_proba + lgbm_proba + cat_proba) / 3
final_pred = np.argmax(final_proba, axis=1)

print(roc_auc_score(final_pred,y_test))
```
